```python
import tensorflow
import keras
import numpy as np
import keras.datasets.mnist as mnist
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers import Flatten
from keras.preprocessing.image import ImageDataGenerator
import matplotlib


np.random.seed(3)
```

# 다중 클래스 ( 네모, 세모, 동그라미) 분류 문제

```python
#1 객체생성 : ImageDataGenerator(객체 부풀리기 성능이 매우 좋음)
# flow_from_directory는 아마 스트림을 만드는 거인듯. 
# flow_from_directory(이미지 경로, target_size={패치이미지크기}, batch_size={배치크기}, class_mode={분류방식: categorical, binary, sparse, None}})


# 3. 데이터셋 생성하기
train_datagen = ImageDataGenerator(rescale=1. /255)
train_generator = train_datagen.flow_from_directory(
    'data/handwriting_shape/train',
    target_size=(24,24),
    batch_size=3,
    class_mode='categorical'
)

test_datagen = ImageDataGenerator(rescale=1. /255)
test_generator = train_datagen.flow_from_directory(
    'data/handwriting_shape/test',
    target_size=(24,24),
    batch_size=3,
    class_mode='categorical'
)
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3,3),
                activation='relu',
                input_shape=(24,24,3)))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(3, activation='softmax'))

from IPython.display import SVG
from IPython.display import display
from PIL import Image
from keras.utils.vis_utils import model_to_dot

%matplotlib inline


SVG(model_to_dot(model, show_shapes=True).create(prog='dot', format='svg'))



```

```python
#모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#6. 모델학습시키기 ( fit 대신 fit_generator를 씀)
model.fit_generator(
    train_generator,
    steps_per_epoch=15,
    epochs=50,
    validation_data=test_generator,
    validation_steps=5
)
```

```python
#5. 모델 평가하기
print( "### Evaluate ###")
scores = model.evaluate_generator(test_generator, steps=5)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]* 100))
```

```python
print('---Predict---')
output= model.predict_generator(test_generator, steps=5)
np.set_printoptions(formatter={'float': lambda x: "{0:0.3f}".format(x)})
print(test_generator.class_indices)
print(output)
```
